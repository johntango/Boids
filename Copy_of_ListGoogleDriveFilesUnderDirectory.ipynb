```python
from google.colab import auth
from googleapiclient.discovery import build
from google.auth.transport.requests import Request
from google.auth import default

# Authenticate user
auth.authenticate_user()
creds, _ = default()

# Initialize Google Drive API client
service = build('drive', 'v3', credentials=creds)
```

```python
# Replace this with your target folder ID
FOLDER_ID = "1OfGtgPvmA97NcFYQWqX9Syz2ZQgAHBwz"

def list_drive_files_in_folder(folder_id):
    query = f"'{folder_id}' in parents"  # Filter by parent folder
    results = service.files().list(
        q=query,
        pageSize=50,
        fields="files(id, name, webViewLink)"
    ).execute()

    files = results.get('files', [])

    if not files:
        print("No files found in the specified folder.")
        return

    for file in files:
        print(f"Name: {file['name']}, Link: {file['webViewLink']}")

# Run the function with the specified folder
list_drive_files_in_folder(FOLDER_ID)
```

```python
import pandas as pd

# Replace these with your folder IDs
MP4_FOLDER_ID = "1AnK_UzuewAx1bGzODsMtYLGyV6wh8xnk"
CMPROJ_FOLDER_ID = "1OfGtgPvmA97NcFYQWqX9Syz2ZQgAHBwz"

def get_drive_files(folder_id, file_extension):
    """Retrieve a list of files from Google Drive with a specific extension."""
    query = f"'{folder_id}' in parents and name contains '{file_extension}'"
    results = service.files().list(
        q=query,
        pageSize=100,
        fields="files(id, name, webViewLink)"
    ).execute()

    files = results.get('files', [])
    return {file['name'][:8]: {"name": file['name'], "link": file['webViewLink']} for file in files}

# Fetch files from both folders
mp4_files = get_drive_files(MP4_FOLDER_ID, ".mp4")
cmproj_files = get_drive_files(CMPROJ_FOLDER_ID, ".cmproj")

# Generate <mit-submodule> entries
submodules = []
for key, mp4 in mp4_files.items():
    cmproj = cmproj_files.get(key, {"name": "", "link": ""})  # Default if no matching cmproj

    submodule = f"""
    <mit-submodule
      title="{mp4['name']}"
      video="{mp4['name']}"
      videoUrl="{mp4['link']}"
      camtasia="{cmproj['name']}"
      camtasiaUrl="{cmproj['link']}"
    ></mit-submodule>
    """

    submodules.append(submodule)

# Print all generated submodules
for submodule in submodules:
    print(submodule)
```
